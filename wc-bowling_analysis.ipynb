```python
# function to get the stats of highest runs, wickets, 4s ,6s, maiden in one match of the tournament. 

def maxInOneMatch(req):
    
    # retrieve the max row according to required subject.
    max_row = df.loc[df[req].idxmax()]
    
    player = max_row['player']
    team = max_row['team']
    req = max_row[req]
    opponent = max_row['opponent']

    return {
        'player':player,
        'team':team,
        'req':req,
        'opponent':opponent
    }
```

# Highest run given by which player in one match?


```python
max_runs = maxInOneMatch('runs')
print(f"Highest runs conceded by {max_runs['player']} from {max_runs['team']} in one of the matches which was {max_runs['req']} runs against {max_runs['opponent']}. ")
```

# Highest wicket taken by which player in one match?


```python
max_wickets = maxInOneMatch('wickets')
print(f"Highest wickets taken by {max_wickets['player']} from {max_wickets['team']} in one of the matches which was {max_wickets['req']} wickets against {max_wickets['opponent']}. ")
```

# Highest 4's given by which player in one match?


```python
max_4s = maxInOneMatch('4s')
print(f"Highest 4s conceded by {max_4s['player']} from {max_4s['team']} in one of the matches which was {max_4s['req']} 4s against {max_4s['opponent']}. ")
```

# Highest 6's given by which player in one match?


```python
max_6s = maxInOneMatch('6s')
print(f"Highest 6s conceded by {max_6s['player']} from {max_6s['team']} in one of the matches which was {max_6s['req']} 6s against {max_6s['opponent']}. ")
```

# Highest maidens bowled by which player in one match?


```python
max_maidens = maxInOneMatch('maidens')
print(f"Highest maidens bowled by {max_maidens['player']} from {max_maidens['team']} in one of the matches which was {max_maidens['req']} maidens against {max_maidens['opponent']}. ")
```

# Highest wicket taker of the tournament?

```python
wicket_takers = df.groupby('player')['wickets'].sum().sort_values(ascending=False).reset_index()
print(f"Highest wicket taker was {wicket_takers.iloc[0].player}, he took {wicket_takers.iloc[0].wickets} wickets.")
```

# Top 5 highest wicket takers?

```python
top_five = wicket_takers[:5]
plt.figure(figsize=(10,5))
sns.barplot(data=top_five, x='player', y='wickets', palette='viridis')
plt.title('Top 5 highest wicket takers')
plt.xlabel('Players')
plt.ylabel('Wickets')
plt.show()
```

```python
# group plyers by overs, runs, wickets, maiden, run-rates..........
df_player_overs = df.groupby('player').agg({'overs':'sum', 'maidens':'sum', 'runs':'sum', 'wickets':'sum','run_rate':'mean', '4s':'sum', '6s':'sum', 'wd':'sum'})
```

```python
df_players = df_player_overs.sort_values(by='overs', ascending=False).reset_index()
```

```python
df_players
```

# Top 10 bowlers with best economy rate & have bowled atleast 50 overs?

```python
temp = df_players[df_players['overs']>=50].sort_values(by='run_rate').reset_index()
best_economy = temp[:10]
```

```python
best_economy
```

```python
plt.figure(figsize=(8,4))
plt.plot(best_economy['player'], best_economy['run_rate'], linestyle='--', marker='o', color='red')
plt.grid()
plt.xticks(rotation=90)
plt.title('Bowlers with best economy rate')
plt.xlabel('Players')
plt.ylabel('Economy rate')
plt.show()
```

# Top 30 bowlers conceded most runs.

```python
temp = df_players.sort_values(by='runs', ascending=False).reset_index()
```

```python
most_runs_conceded = temp[:30]
```

```python
plt.figure(figsize=(12,4))
sns.barplot(data=most_runs_conceded, x='player', y='runs', palette='viridis')
plt.title('Most runs conceded by bowlers')
plt.xticks(rotation=90)
plt.show()
```
